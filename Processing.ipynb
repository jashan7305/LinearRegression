```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import joblib
```

```python
df = pd.read_csv("data\\Admission_Predict.csv")
df.drop(columns=["Serial No."], inplace=True)
df.head()
```

```python
scaler = StandardScaler()
df[["GRE Score", "TOEFL Score", "SOP", "LOR ", "CGPA"]] = scaler.fit_transform(df[["GRE Score", "TOEFL Score", "SOP", "LOR ", "CGPA"]])
df.head()
```

```python
df = pd.get_dummies(data=df, columns=["University Rating", "Research"])
df.head()
```

```python
joblib.dump(scaler, "models\\scaler.pkl")
df.to_csv("data\\data.csv", index=False)
```
